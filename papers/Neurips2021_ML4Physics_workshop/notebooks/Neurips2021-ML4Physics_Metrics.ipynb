# Plot: RMSE vs wavelength

```python
def define_plot_style():
    # Define plot paramters
    # Use seaborn style
    sns.set()
    
    plot_style = {
        'figure.figsize': (12,8),
        'figure.dpi': 200,
        'figure.autolayout':True,
        'lines.linewidth': 2,
        'lines.linestyle': '-',
        'lines.marker': 'o',
        'lines.markersize': 10,
        'legend.fontsize': 20,
        'legend.loc': 'best',
        'axes.titlesize': 24,
        'font.size': 18
    }
    mpl.rcParams.update(plot_style)

saving_dir_path = '/Users/tliaudat/Documents/PhD/myPapers/2021-instrumental_response-Neurips_ML4Physics/plots/'
```

```python
wf_dd_metrics['test_metrics']['mono_metric'].keys()
```

```python
define_plot_style()
lambda_list = np.arange(0.55, 0.9, 0.01)

metrics = wf_dd_metrics

sns.set(font_scale=1.3)


## Monochromatic 
fig = plt.figure(figsize=(14,8))
ax1 = fig.add_subplot(111)

ax1.errorbar(
    x = lambda_list,
    y = metrics['test_metrics']['mono_metric']['rmse_lda'],
    yerr = metrics['test_metrics']['mono_metric']['std_rmse_lda'],
    label='Model (iii) Zernike 15 + DD',
    alpha=0.75
)
# ax1.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.1e'))
plt.minorticks_on()
# plt.ticklabel_format(axis=ax1, style="sci", scilimits=(0,0))
# ax1.yaxis.formatter.set_powerlimits((0, 0))
ax1.yaxis.major.formatter._useMathText = True
ax1.ticklabel_format(style='sci',scilimits=(0,0),axis='y')
ax1.yaxis.set_offset_position('left')
ax1.yaxis.offsetText.set(size=16)


ax1.legend(prop={'size': 18})

ax1.set_xlabel('Wavelength [um]', fontsize = 18)
ax1.set_ylabel('RMSE', fontsize = 18)

ax2 = ax1.twinx()
kwargs = dict(linewidth=2, linestyle='dashed', markersize=8, marker='^', alpha=0.5)

ax2.plot(
    lambda_list,
    metrics['test_metrics']['mono_metric']['rel_rmse_lda'],
    **kwargs
)
ax2.set_ylabel('Relative RMSE [%]', fontsize = 18)  
ax2.grid(False)

plt.minorticks_on()

plt.savefig(saving_dir_path + 'model_iii_z15_DD_monochrom_pixel_RMSE_v2.pdf')

plt.show()
        
```
